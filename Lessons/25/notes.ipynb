# Regex - wyrażenia regularne

## Wyrażenia regularne (regex)

Specjalne wzorce, które umożliwiają przeszukiwanie, modyfikowanie oraz analizowanie tekstu na podstawie zadanych reguł.

Zastosowania wyrażeń regularnych:
- wyszukiwanie tekstu,
- weryfikacja formatu danych,
- zamiana tekstu,
- parsowanie i ektrakcja danych,
- podział tekstu.

Przydatne narzędzie do analizowania wzorców - https://regex101.com

## Biblioteka `re`

`re` to standardowa biblioteka w Pythonie służąca do pracy z wyrażeniami regularnymi (regex).
Wyrażenia regularne to narzędzie do przeszukiwania, dopasowywania i manipulacji tekstu na podstawie zdefiniowanych wzorców.
Biblioteka re oferuje zestaw funkcji i metod, które pozwalają na wykonywanie takich operacji jak wyszukiwanie wzorców w tekście, ich zamiana, czy podział tekstu na części.

[Dokumentacja biblioteki `re`](https://docs.python.org/3/library/re.html)

Najpopularniejsze funkcje biblioteki `re`:
- `re.search(pattern, string)` - Szuka pierwszego wystąpienia wzorca w tekście. Zwraca obiekt `Match` lub `None` jeśli nie znajdzie pasującego fragmentu.
- `re.match(pattern, string)` - Sprawdza, czy wzorzec pasuje na początku tekstu. Zwraca obiekt `Match` lub `None`.
- `re.findall(pattern, string)` - Zwraca listę wszystkich fragmentów tekstu, które pasują do wzorca.
- `re.sub(pattern, repl, string)` - Zastępuje wszystkie wystąpienia wzorca w tekście na podany ciąg znaków `repl`.

## Wzorzec

Wzorzec (ang. pattern) to specjalny ciąg znaków, który definiuje reguły wyszukiwania i dopasowywania fragmentów tekstu.
Wzorzec jest tworzony za pomocą specjalnych symboli, znaków i metaznaków, które określają, co powinno być dopasowane w tekście. Może być to np. konkretny ciąg znaków, pewien zakres liter lub cyfr, liczba powtórzeń określonego fragmentu czy dowolny znak. Wzorzec definiuje, jakiego rodzaju dane chcemy znaleźć w tekście.

Kluczowe elementy wzorców:
- znaki literowe i specjalne - np. `abc`,
- metaznaki - np. `.`, która oznacza jeden dowolny znak z wyjątkiem znaku nowej linii,
- kwantyfikatory - np. `*`, która oznacza dopasowanie zero lub więcej wystąpień poprzedniego znaku lub wyrażenia,
- klasa znaków - np. `[abc]` oznaczające, że dowolny z wymienionych znaków (`a`, `b` lub `c`) będzie pasować,
- grupowanie i alternatywa - `()` używane do tworzenia podwyrażeń i `|` oznaczające "albo",
- znaki specjalne - znaki takie jak dosłowna kropka (`.`) trzeba poprzedzić `\`, ponieważ mają specjalne znaczenie w regex, na przykład `\.` dopasuje kropkę, a nie dowolny znak.

## Flagi

Flagi w wyrażeniach regularnych są opcjami, które wpływają na sposób interpretacji wzorca przez funkcje z biblioteki `re`.
Dzięki flagom można dostosować sposób, w jaki wzorzec wyszukuje i dopasowuje tekst, co czyni operacje z wyrażeniami regularnymi bardziej elastycznymi i dopasowanymi do różnych potrzeb.

Najważniejsze flagi:
- `re.IGNORECASE` (`re.I`) - Umożliwia ignorowanie wielkości liter podczas dopasowywania wzorca. Dzięki tej fladze wzorzec będzie pasował zarówno do małych, jak i wielkich liter.
- `re.MULTILINE` (`re.M`) - Zmienia sposób, w jaki wzorzec traktuje znaki `^` (początek) i `$` (koniec). Zwykle `^` dopasowuje początek całego ciągu, a `$` jego koniec. Z flagą `re.MULTILINE`, `^` i `$` będą dopasowywać odpowiednio początek i koniec każdej linii w wielowierszowym tekście.
- `re.DOTALL` (`re.S`) - Zmienia działanie kropki `.` w wyrażeniach regularnych. Domyślnie `.` dopasowuje dowolny znak z wyjątkiem znaku nowej linii `\n`. Z flagą `re.DOTALL`, kropka dopasowuje także znaki nowej linii.
- `re.VERBOSE` (`re.X`) - Umożliwia pisanie wzorców w bardziej czytelny sposób. Dzięki tej fladze, można wprowadzać białe znaki (spacje, tabulacje) i komentarze wewnątrz wzorca, które będą ignorowane. To ułatwia pisanie i zrozumienie bardziej złożonych wyrażeń regularnych.
- `re.ASCII` (`re.A`) - Sprawia, że wzorzec będzie dopasowywał wyłącznie znaki ASCII. Ma to znaczenie przy stosowaniu skrótów takich jak `\w`, `\d`, `\s`, które z flagą `re.ASCII` będą dopasowywać tylko odpowiednie znaki ASCII, a nie np. litery diakrytyczne spoza zestawu ASCII.

## Funkcja `re.match`

```python
sentence = "Ala ma kota"
result = re.match(r"Ala", "Ala ma kota")
print(f"Wyszukiwanie elementu `Ala`: {result}")
result = re.match(r"kota", "Ala ma kota")
print(f"Wyszukiwanie elementu `kota`: {result}")
```

```python
sentence = "Ala ma kota"
result = re.match(r"Ala", sentence)
print(f"Wyszukiwanie elementu `Ala`: {result}")
print(f"Znaleziono element `{result.group()}`")
print(f"Poczatek dopasowania znajduje się pod indeksem {result.start()}")
print(f"Koniec dopasowania następuje pod indeksem {result.end()}")
print(f"Krotka zawierająca przedział z dopasowaniem {result.span()}")
```

```python
result = re.match(r"ala", sentence)
print(f"Wyszukiwanie elementu `ala`: {result}")
```

```python
result = re.match(r"ala", sentence, re.IGNORECASE)
print(f"Wyszukiwanie elementu `ala`: {result}")
```

## Funkcja `re.search`

```python
napis = "Ala ma 123 jabłka"
result = re.search(r"\d+", napis)
print(f"Wyszukiwanie liczby zakończone, wykryto: {result}")
```

## Funkcja `re.findall`

```python
napis = "Ala i Kuba jadą na wycieCzkę do Warszawy"
result = re.findall(r"\b[A-Z][a-z]+", napis)
print(f"Wyszukiwanie wyrazów zaczynających się wielką literą zakończone, wykryto: {result}")
```

## Funkcja `re.sub`

```python
text = "Valorant to najlepsza strzelanka na świecie"
pattern = "Valorant"
replacement = "CS"
new_text = re.sub(pattern, replacement, text)
print(new_text)
```
