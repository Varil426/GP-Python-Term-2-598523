## Funkcje anonimowe

Funkcje anonimowe, zwane również funkcjami lambda, są jednowierszowymi funkcjami nie wymagającymi słowa
kluczowego `def`. Idealnie nadają się do prostych operacji, szczególnie w kontekście przetwarzania danych,
sortowania, czy filtracji.

```python
def add(a, b):
    return a + b


print(f"Function: {add(2, 3)}")

lambda_add = lambda x, y: x + y
print(f"Lambda: {lambda_add(2, 3)}")
```

```python
example_list = [x for x in range(100)]

divisible_by_12 = filter(lambda x: x % 12 == 0, example_list)
print(list(divisible_by_12))
```

## `*args` i `**kwargs`

Funkcje w Pythonie nie muszą posiadać odgórnie zdefiniowanej listy przyjmowanych argumentów. `*args` i
`**kwargs` to mechanizmy pozwalające na przekazywanie do funkcji zmiennej liczby argumentów.

`*args` – Niepozycjonowane argumenty
- Używane do przekazywania dowolnej liczby niepozycjonowanych argumentów do funkcji.
- Przekazane argumenty są zbierane w formie krotki (`tuple`).

`**kwargs` – Słownik z argumentami nazwanymi
- Używane do przekazywania dowolnej liczby nazwanych argumentów (klucz-wartość) do funkcji.
- Przekazane argumenty są zbierane w formie słownika (`dict`).

```python
def sum_numbers(*args):
    s = 0
    for num in args:
        s += num
    return s

print(f"Sum: {sum_numbers(1, 2, 3, 4)}")
```

```python
def print_city_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("-"*40)
print_city_details(name="New York", population=8000000, area=468.9)
print("-"*40)
print_city_details(name="Los Angeles", population=4000000, area=503, country="USA", currency="USD")
```

## Generatory

Generatory w Pythonie to specjalny rodzaj funkcji, które tworzą sekwencje wartości w
sposób stopniowy, zamiast tworzyć całą listę od razu. To oznacza, że generatory pozwalają
oszczędzać pamięć i są bardziej wydajne, gdy pracujesz z dużymi zestawami danych.

```python
def get_values():
    while True:
        yield 100
        yield 200
        yield 300

generator = get_values()

for _ in range(10):
    print(next(generator))
```

## Dekoratory

Dekoratory to specjalne funkcje w Pythonie, które pozwalają modyfikować lub rozszerzać
zachowanie innej funkcji, klasy lub metody, bez potrzeby zmiany ich kodu źródłowego. Są
często używane, aby uczynić kod bardziej modularnym, przejrzystym i łatwiejszym w
utrzymaniu.

Dekoratory to funkcje, które przyjmują jako argument inną funkcję (lub klasę) i zwracają
nową funkcję (lub klasę) z dodatkowymi modyfikacjami. W Pythonie stosuje się je za
pomocą symbolu `@`, co sprawia, że kod jest czytelniejszy.

```python
def to_upper(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper

def greet():
    return "Hello, World!"

@to_upper
def greet_decorated():
    return greet()

print(greet())
print(greet_decorated())
```
