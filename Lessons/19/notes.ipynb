# Metody list

## Listy

Listy w Pythonie to wbudowane struktury danych, które służą do przechowywania uporządkowanych kolekcji elementów.
Listy mogą zawierać elementy dowolnego typu.

## Cechy list

1. Dynamiczny rozmiar - listy mogą zmieniać swoją długość w trakcie działania programu. Można dodawać, usuwać i modyfikować elementy listy.
1. Indeksowane - elementy listy są uporządkowane i możne je pobierać przy użyciu indeksów, które zaczynają się od 0.
1. Zmienne - są typem zmiennym, co oznacza, że można zmieniać ich zawartość po ich utworzeniu.
1. Mieszane typy danych - listy mogą przechowywać elementy różnych typów.

## Metody list

```python
lista = [1, "tekst", 3.14, True, None, 42]
print("Początkowa lista:", lista)
```

```python
# append() - dodaje element na końcu listy
print("append(100) - dodaje element na końcu:")
lista.append(100)
print(lista)
```

```python
# extend() - rozszerza listę o elementy z innej listy/iterowalnego obiektu
print("extend([200, 300]) - rozszerza listę o elementy z innej listy:")
lista.extend([200, 300])
print(lista)
```

```python
# insert() - wstawia element na określonej pozycji
print("insert(2, 'nowy') - wstawia element na pozycji 2:")
lista.insert(2, 'nowy')
print(lista)
```

```python
# remove() - usuwa pierwsze wystąpienie podanej wartości
print("remove(True) - usuwa pierwsze wystąpienie wartości True:")
lista.remove(True)
print(lista)
```

```python
# pop() - usuwa i zwraca element z podanej pozycji (domyślnie ostatni)
print("pop() - usuwa i zwraca ostatni element:")
usuniety = lista.pop()
print(f"Usunięty element: {usuniety}")
print(lista)

print("pop(0) - usuwa i zwraca element z pozycji 0:")
pierwszy = lista.pop(0)
print(f"Usunięty element: {pierwszy}")
print(lista)
```

```python
# index() - zwraca indeks pierwszego wystąpienia wartości
print("index(42) - zwraca indeks pierwszego wystąpienia wartości 42:")
indeks = lista.index(42)
print(f"Indeks wartości 42: {indeks}")
```

```python
# count() - zlicza wystąpienia wartości w liście
print("count('tekst') - zlicza wystąpienia wartości 'tekst':")
liczba = lista.count('tekst')
print(f"Liczba wystąpień 'tekst': {liczba}")
```

```python
# sort() - sortuje listę (działa tylko dla list z porównywalnymi elementami)
print("sort() - sortuje listę (tworzymy nową listę z liczbami):")
lista_liczb = [5, 2, 8, 1, 9, 3]
print(f"Przed sortowaniem: {lista_liczb}")
lista_liczb.sort()
print(f"Po sortowaniu: {lista_liczb}")


print("sort(reverse=True) - sortuje malejąco:")
lista_liczb.sort(reverse=True)
print(f"Po sortowaniu malejąco: {lista_liczb}")
```

```python
# reverse() - odwraca kolejność elementów w liście (modyfikuje oryginalną listę)
print("\n--- Demonstracja reverse() ---")
lista.reverse()
print("Po reverse():", lista)

# Ponowne użycie reverse() przywraca oryginalną kolejność
lista.reverse()
print("Po ponownym reverse():", lista)
```

```python
# copy() - tworzy płytką kopię listy (zmiany w kopii nie wpływają na oryginał)
print("\n--- Demonstracja copy() ---")
lista_kopia = lista.copy()
print("Oryginalna lista:", lista)
print("Kopia listy:", lista_kopia)

# Modyfikacja kopii nie wpływa na oryginał
lista_kopia.append("nowy element")
lista_kopia[0] = 999
print("\nPo modyfikacji kopii:")
print("Oryginalna lista:", lista)
print("Zmodyfikowana kopia:", lista_kopia)
```

```python
# clear() - usuwa wszystkie elementy z listy (pozostawia pustą listę)
print("\n--- Demonstracja clear() ---")
lista_do_wyczyszczenia = [10, 20, 30, 40, 50]
print("Lista przed clear():", lista_do_wyczyszczenia)
lista_do_wyczyszczenia.clear()
print("Lista po clear():", lista_do_wyczyszczenia)
print("Długość listy:", len(lista_do_wyczyszczenia))
```
