## Czym jest API? A co to REST?

### API

API (Application Programming Interface) - zestaw zasad i protokołów, które pozwalają różnym programom komunikować się ze sobą.

### REST

REST (Representational State Transfer) - styl architektury, który umożliwia tworzenie API działających na zasadzie prostych zasad i protokołów. API, które jest zgodne z zasadami REST, nazywamy REST API.

Zasady REST:
- Architektura klient-serwer - Zasada oddzielenia klienta i serwera oznacza, że obie strony mogą rozwijać się niezależnie. Klient (np. aplikacja frontendowa) wysyła żądania do serwera (np. API), który przetwarza te żądania i zwraca odpowiedzi.
- Bezstanowość - Każde zapytanie od klienta do serwera musi zawierać wszystkie informacje potrzebne do przetworzenia tego zapytania. Serwer nie przechowuje żadnego stanu aplikacji pomiędzy zapytaniami.
- Jednolity interfejs - Wszystkie zasoby powinny być dostępne pod przewidywalnymi i intuicyjnymi adresami URL, a operacje na zasobach są realizowane za pomocą standardowych metod HTTP (np. `GET`, `POST`, `PUT`, `DELETE`).
- Podział na zasoby - Zasoby w REST są reprezentowane przez unikalne adresy URL (URI).

Podstawowe operacje REST API:
- `GET` - pobieranie danych,
- `POST` - tworzenie danych,
- `PUT` - modyfikowanie danych,
- `DELETE` - usuwanie danych.

## Biblioteka `requests`

`requests` to biblioteka w Pythonie, która służy do wysyłania zapytań HTTP (takich jak `GET`, `POST`, `PUT`, `DELETE`) do serwerów. Jest to narzędzie, które pozwala nam pobierać dane z internetu, wysyłać dane do serwerów i wchodzić w interakcję z różnymi usługami sieciowymi, takimi jak API.

Zalety:
- Prosta w użyciu z intuicyjną składnią - nie trzeba znać szczegółów protokołu HTTP.
- W porównaniu z innymi bibliotekami (np. `urllib`) wymaga mniej kodu i oferuje czytelniejszy sposób na wysyłanie żądań.
- Obsługuje różne rodzaje żądań HTTP (np. `GET`, `POST`, `PUT`, `DELETE`) oraz pozwala na łatwe przekazywanie danych i obsługę odpowiedzi z serwera.

## Headers i body

Nagłówki (headers) to dodatkowe informacje, które są przesyłane wraz z zapytaniem lub odpowiedzią HTTP. Znajdują się one na początku każdego zapytania lub odpowiedzi.

Body (treść) to główna zawartość zapytania lub odpowiedzi, czyli dane, które są faktycznie przesyłane między klientem (np. przeglądarką lub aplikacją) a serwerem.

## Kody odpowiedzi z serwera

Kody odpowiedzi serwera HTTP (HTTP status codes) są liczbami trzycyfrowymi, które informują o wyniku przetwarzania żądania przez serwer. Każdy z kodów należy do jednej z pięciu kategorii, a każda z tych kategorii ma swoje specyficzne znaczenie.

1. 1xx - Informacyjne (Informational)
1. 2xx - Sukces (Success)
1. 3xx - Przekierowania (Redirects)
1. 4xx - Błędy klienta (Client Error)
1. 5xx - Błędy serwera (Server Error)
