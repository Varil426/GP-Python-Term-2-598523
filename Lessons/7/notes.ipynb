```python
# Sprytniejsze rozwiązanie

oceny = []

# Oddaliśmy kontrolę nad liczbą ocen użytkownikowi
ile_ocen = int(input("Ile ocen chcesz podać? "))

i = 0
while i < ile_ocen:
    ocena = input("Podaj ocenę: ")
    oceny.append(ocena)
    i += 1
```

## Instrukcja `break`

```python
import time

while True:
    print("You can't stop me now!")
    time.sleep(1)
    break

print("Haa! Maybe I - the programmer - can!")
```

```python
print("Będę Twoim echo")

while True:
    odp = input("Napisz coś: ")
    if odp == "exit":
        break
    print("Echo: " + odp)

print("Do zobaczenia!")
```

## Instrukcja `continue`

```python
PIN = "1234"
ROK_URODZENIA = "1998"
HASLO = "qwerty"

while True:
    input_pin = input("PIN: ")
    if input_pin != PIN:
        print("Odmowa dostępu!")
        continue

    input_rok_urodzenia = input("Rok urodzenia: ")
    if input_rok_urodzenia != ROK_URODZENIA:
        print("Odmowa dostępu! Zaloguj się ponownie")
        continue

    input_haslo = input("Hasło: ")
    if input_haslo != HASLO:
        print("Odmowa dostępu! Zaloguj się ponownie")
        continue

    print("Zalogowano poprawnie!")
    break

print("Super tajne treści znajdują się tutaj")
```
