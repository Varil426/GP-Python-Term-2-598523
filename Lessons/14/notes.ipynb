## Stack, stos, LIFO



```python
stack = []

stack.append('A')
stack.append('B')
stack.append('C')

print("Stos po dodaniu elementów:", stack)

element = stack.pop()

print("Usunięty element:", element)
print("Stos po usunięciu elementu:", stack)
```
