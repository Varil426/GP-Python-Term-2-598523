## Protokół SMTP

SMTP, czyli Simple Mail Transfer Protocol, to prosty protokół służący do wysyłania e-maili przez Internet.
Wyobraź sobie, że to taki listonosz, który odbiera Twój e-mail i dostarcza go do skrzynki odbiorcy.
Jest to standardowy sposób, w jaki komputery "rozmawiają", żeby przesyłać wiadomości e-mail.

Jak działa SMTP?
1. Nadawca przygotowuje wiadomość.
1. Wiadomość jest przesyłana na serwer pocztowy za pomocą SMTP.
1. Serwer SMTP określa odbiorcę i wysyła do niego wiadomość.
1. Odbiorca odbiera wiadomość.

Zalety SMTP:
- Uniwersalność
- Prostota
- Szybkość
- Stabilność

## Plik `.env`

Plik `.env` to zwykły plik tekstowy, który zawiera pary klucz-wartość, przechowujące zmienne
środowiskowe używane przez aplikację. Jego głównym celem jest przechowywanie wrażliwych danych
konfiguracyjnych w jednym miejscu, w sposób bezpieczny i łatwy w zarządzaniu.

Co przechowujemy w pliku `.env`?
- Hasła
- Tokeny API
- Adresy serwerów
- Ustawienia konfiguracyjne

Dlaczego używamy plików `.env`?
- Oddzielenie kodu od danych konfiguracyjnych
- Bezpieczeństwo
- Łatwość zarządzania konfiguracją
- Przenośność
